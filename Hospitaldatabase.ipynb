```python
import mysql.connector

# ================= MYSQL CONNECTION =================
conn = mysql.connector.connect(
    host="localhost",
    user="root",      
    password="Root@123#"     
)
cursor = conn.cursor()

# ================= CREATE DATABASE =================
cursor.execute("CREATE DATABASE IF NOT EXISTS ClinicDB")
cursor.execute("USE ClinicDB")

# ================= CREATE TABLES =================

cursor.execute("""
CREATE TABLE IF NOT EXISTS PatientMaster (
    PID INT AUTO_INCREMENT PRIMARY KEY,
    FullName VARCHAR(100),
    DOB DATE,
    Gender ENUM('Male','Female'),
    City VARCHAR(50),
    Phone VARCHAR(15) UNIQUE
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS Unit (
    UnitID INT AUTO_INCREMENT PRIMARY KEY,
    UnitName VARCHAR(100),
    Floor INT
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS MedicalStaff (
    StaffID INT AUTO_INCREMENT PRIMARY KEY,
    StaffName VARCHAR(100),
    Role VARCHAR(50),
    UnitID INT,
    FOREIGN KEY (UnitID) REFERENCES Unit(UnitID)
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS Visit (
    VisitID INT AUTO_INCREMENT PRIMARY KEY,
    PID INT,
    StaffID INT,
    VisitDate DATE,
    Notes VARCHAR(200),
    FOREIGN KEY (PID) REFERENCES PatientMaster(PID),
    FOREIGN KEY (StaffID) REFERENCES MedicalStaff(StaffID)
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS Treatment (
    TreatmentID INT AUTO_INCREMENT PRIMARY KEY,
    VisitID INT,
    Description VARCHAR(150),
    DurationDays INT,
    FOREIGN KEY (VisitID) REFERENCES Visit(VisitID)
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS Invoice (
    InvoiceID INT AUTO_INCREMENT PRIMARY KEY,
    PID INT,
    Amount DECIMAL(8,2),
    PaymentStatus VARCHAR(30) DEFAULT 'Pending',
    FOREIGN KEY (PID) REFERENCES PatientMaster(PID)
)
""")

conn.commit()
cursor.close()
conn.close()

print("‚úÖ Database and tables created successfully")
```

```python
import mysql.connector
import random
from datetime import datetime, timedelta

# ================= MYSQL CONNECTION =================
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Root@123#",
    database="HospitalDB"
)
cursor = conn.cursor()

# ================= RANDOM DATA =================
male_names = ["Aman", "Rohit", "Kunal", "Arjun", "Siddharth"]
female_names = ["Neha", "Anjali", "Pooja", "Riya", "Kavya"]
cities = ["Raipur", "Delhi", "Indore", "Bhopal", "Nagpur"]

def random_dob():
    start = datetime(1970, 1, 1)
    end = datetime(2005, 12, 31)
    return start + timedelta(days=random.randint(0, (end-start).days))

# ================= INSERT DATA =================
for i in range(100):
    gender = random.choice(["Male", "Female"])
    name = random.choice(male_names if gender == "Male" else female_names)
    phone = "9" + str(random.randint(100000000, 999999999))
    
    cursor.execute("""
        INSERT INTO PatientMaster (FullName, DOB, Gender, City, Phone)
        VALUES (%s, %s, %s, %s, %s)
    """, (
        name + " Verma",
        random_dob().date(),
        gender,
        random.choice(cities),
        phone
    ))

conn.commit()
cursor.close()
conn.close()

print("‚úÖ 100 random patient records inserted successfully")
```
