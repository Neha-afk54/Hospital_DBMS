```python
import mysql.connector

# -----------------------------
# MySQL Connection
# -----------------------------
conn = mysql.connector.connect(
    host="localhost",
    user="root",          
    password="Root@123#" 
)

cursor = conn.cursor()

# -----------------------------
# Create Database
# -----------------------------
cursor.execute("CREATE DATABASE IF NOT EXISTS HospitalManagement")
cursor.execute("USE HospitalManagement")

# -----------------------------
# Patient Table
# -----------------------------
cursor.execute("""
CREATE TABLE IF NOT EXISTS Patient (
    PatientID INT PRIMARY KEY,
    Name VARCHAR(100) NOT NULL,
    DateOfBirth DATE,
    Gender VARCHAR(10),
    Address VARCHAR(255),
    PhoneNumber VARCHAR(15),
    Email VARCHAR(100)
)
""")

# -----------------------------
# Department Table
# -----------------------------
cursor.execute("""
CREATE TABLE IF NOT EXISTS Department (
    DepartmentID INT PRIMARY KEY,
    DepartmentName VARCHAR(100) NOT NULL,
    Location VARCHAR(100)
)
""")

# -----------------------------
# Doctor Table
# -----------------------------
cursor.execute("""
CREATE TABLE IF NOT EXISTS Doctor (
    DoctorID INT PRIMARY KEY,
    Name VARCHAR(100) NOT NULL,
    Specialization VARCHAR(100),
    PhoneNumber VARCHAR(15),
    Email VARCHAR(100),
    DepartmentID INT,
    FOREIGN KEY (DepartmentID) REFERENCES Department(DepartmentID)
)
""")

# -----------------------------
# Appointment Table
# -----------------------------
cursor.execute("""
CREATE TABLE IF NOT EXISTS Appointment (
    AppointmentID INT PRIMARY KEY,
    PatientID INT,
    DoctorID INT,
    AppointmentDate DATE,
    AppointmentTime TIME,
    Status VARCHAR(50),
    FOREIGN KEY (PatientID) REFERENCES Patient(PatientID),
    FOREIGN KEY (DoctorID) REFERENCES Doctor(DoctorID)
)
""")

# -----------------------------
# Prescription Table
# -----------------------------
cursor.execute("""
CREATE TABLE IF NOT EXISTS Prescription (
    PrescriptionID INT PRIMARY KEY,
    AppointmentID INT,
    MedicationDetails TEXT,
    Dosage VARCHAR(50),
    Duration VARCHAR(50),
    FOREIGN KEY (AppointmentID) REFERENCES Appointment(AppointmentID)
)
""")

# -----------------------------
# Billing Table
# -----------------------------
cursor.execute("""
CREATE TABLE IF NOT EXISTS Billing (
    BillID INT PRIMARY KEY,
    PatientID INT,
    Amount DECIMAL(10,2),
    BillDate DATE,
    PaymentStatus VARCHAR(50),
    FOREIGN KEY (PatientID) REFERENCES Patient(PatientID)
)
""")

conn.commit()
conn.close()

print("MySQL Hospital Management Database created successfully!")
```

```python
import mysql.connector
import random
from datetime import date, time

# -----------------------------
# MySQL Connection
# -----------------------------
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Root@123#",
    database="HospitalManagement"
)

cursor = conn.cursor()

# -----------------------------
# Indian Sample Data
# -----------------------------
patient_names = [
    "Neha", "Riya", "Sidharth", "Vikash", "Ankit", "Pooja", "Rahul",
    "Aman", "Sneha", "Karan", "Priya", "Rohit", "Shreya", "Nikhil"
]

doctor_names = [
    "Dr. Amit Sharma", "Dr. Sunita Verma", "Dr. Rajesh Gupta",
    "Dr. Neeraj Singh", "Dr. Kavita Rao", "Dr. Suresh Patel"
]

specializations = [
    "Cardiology", "Neurology", "Orthopedics",
    "Pediatrics", "General Medicine", "Gynecology"
]

medicines = [
    "Paracetamol", "Azithromycin", "Ibuprofen",
    "Amoxicillin", "Vitamin D", "Cough Syrup"
]

cities = ["Raipur", "Delhi", "Mumbai", "Pune", "Bhopal", "Nagpur"]

# -----------------------------
# Insert Departments
# -----------------------------
departments = [
    (1, "Cardiology", "Block A"),
    (2, "Neurology", "Block B"),
    (3, "Orthopedics", "Block C"),
    (4, "Pediatrics", "Block D"),
    (5, "General Medicine", "Block E")
]

cursor.executemany(
    "INSERT IGNORE INTO Department VALUES (%s,%s,%s)",
    departments
)

# -----------------------------
# Insert Patients (1000)
# -----------------------------
patients = []
for i in range(1, 1001):
    name = random.choice(patient_names)
    patients.append((
        i,
        name,
        date(random.randint(1980, 2010), random.randint(1, 12), random.randint(1, 28)),
        random.choice(["Male", "Female"]),
        f"{random.choice(cities)}, India",
        f"9{random.randint(100000000,999999999)}",
        f"{name.lower()}{i}@gmail.com"
    ))

cursor.executemany(
    "INSERT IGNORE INTO Patient VALUES (%s,%s,%s,%s,%s,%s,%s)",
    patients
)

# -----------------------------
# Insert Doctors
# -----------------------------
doctors = []
for i in range(1, 51):
    doctors.append((
        i,
        random.choice(doctor_names),
        random.choice(specializations),
        f"8{random.randint(100000000,999999999)}",
        f"doctor{i}@hospital.in",
        random.randint(1, 5)
    ))

cursor.executemany(
    "INSERT IGNORE INTO Doctor VALUES (%s,%s,%s,%s,%s,%s)",
    doctors
)

# -----------------------------
# Insert Appointments
# -----------------------------
appointments = []
for i in range(1, 1001):
    appointments.append((
        i,
        random.randint(1, 1000),
        random.randint(1, 50),
        date(2025, random.randint(1, 12), random.randint(1, 28)),
        time(random.randint(9, 17), 0),
        random.choice(["Scheduled", "Completed"])
    ))

cursor.executemany(
    "INSERT IGNORE INTO Appointment VALUES (%s,%s,%s,%s,%s,%s)",
    appointments
)

# -----------------------------
# Insert Prescriptions
# -----------------------------
prescriptions = []
for i in range(1, 1001):
    prescriptions.append((
        i,
        i,
        random.choice(medicines),
        "1 Tablet Twice a Day",
        "5 Days"
    ))

cursor.executemany(
    "INSERT IGNORE INTO Prescription VALUES (%s,%s,%s,%s,%s)",
    prescriptions
)

# -----------------------------
# Insert Billing
# -----------------------------
billing = []
for i in range(1, 1001):
    billing.append((
        i,
        i,
        random.randint(500, 8000),
        date(2025, random.randint(1, 12), random.randint(1, 28)),
        random.choice(["Paid", "Unpaid"])
    ))

cursor.executemany(
    "INSERT IGNORE INTO Billing VALUES (%s,%s,%s,%s,%s)",
    billing
)

conn.commit()
conn.close()

print("âœ… 1000 Indian dummy records inserted successfully!")
```

```python
import mysql.connector
import pandas as pd
import os
import time

# -----------------------------
# MySQL Connection
# -----------------------------
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Root@123#",
    database="HospitalManagement"
)

# -----------------------------
# Fetch ALL 1000 Records
# -----------------------------
query = """
SELECT 
    PatientID AS PID,
    Name AS FullName,
    DateOfBirth AS DOB,
    Gender,
    Address AS City,
    PhoneNumber AS Phone,
    Email 
FROM Patient
ORDER BY PatientID
"""

df = pd.read_sql(query, conn)

# -----------------------------
# FORCE DISPLAY ALL 1000 ROWS
# -----------------------------
pd.set_option("display.max_rows", 1000)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 150)

print("\nðŸ“‹ FULL PATIENT TABLE (1000 RECORDS)")
display(df)

# -----------------------------
# Gender Distribution
# -----------------------------
print("\nðŸ“Š Gender Distribution")
print(df["Gender"].value_counts())

# -----------------------------
# EXPORT TO EXCEL (ALL 1000)
# -----------------------------
folder_path = r"C:\Users\sonka\Documents\DBMS"
os.makedirs(folder_path, exist_ok=True)

timestamp = time.strftime("%Y%m%d_%H%M%S")
excel_path = f"{folder_path}/ClinicDB_PatientData_{timestamp}.xlsx"

df.to_excel(excel_path, index=False)

print(f"\nðŸ“ Excel file created successfully at:\n{excel_path}")

conn.close()
```
